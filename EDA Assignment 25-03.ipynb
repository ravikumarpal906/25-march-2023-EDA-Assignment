# **Features to analyze:**

* **Destination:** The destination city of the flight can also be a factor in flight price trends. Some destinations may be more popular during certain times of the year, leading to higher prices.

* **Line chart:** A line chart can be used to visualize the average flight price for each month over time. This can help to identify any seasonal trends in flight prices.

# **Example visualization:**

The following code creates a line chart that visualizes the average flight price for each month over time:


```python
# Average price trend over time
monthly_avg_price = flight_df.groupby('Month')['Price'].mean().reset_index()
plt.plot(monthly_avg_price['Month'], monthly_avg_price['Price'])
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.title('Average Flight Price Trend by Month')
plt.show()

# Average price comparison by airline
airline_avg_price = flight_df.groupby('Airline')['Price'].mean().reset_index()
plt.bar(airline_avg_price['Airline'], airline_avg_price['Price'])
plt.xlabel('Airline')
plt.ylabel('Average Price')
plt.title('Average Flight Price Comparison by Airline')
plt.xticks(rotation=45)
plt.show()
```

**Q11. Are there any missing values in the dataset? Identify any missing values and describe how they may impact your analysis.**

```python
# Check for missing values in each column
missing_values = playstore_df.isnull().sum()

# Identify columns with missing values
columns_with_missing_values = missing_values[missing_values > 0].index.tolist()

# Describe how missing values may impact the analysis
for column in columns_with_missing_values:
  print(f"Missing values in column '{column}': {missing_values[column]}")
  print("Impact on analysis:")
  if column == 'Rating':
    print("- Ratings are an important factor in determining the quality of an app.")
    print("- Missing ratings can make it difficult to compare apps and identify the best ones.")
  elif column == 'Reviews':
    print("- Reviews provide valuable feedback from users about their experiences with the app.")
    print("- Missing reviews can limit the ability to understand user sentiment and identify areas for improvement.")
  # Add specific analysis impact for other relevant columns
  print()
```

**Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.**

```python
playstore_df['Size'] = playstore_df['Size'].str.replace('M','000')
playstore_df['Size'] = playstore_df['Size'].str.replace('k','')
playstore_df['Size'] = playstore_df['Size'].str.replace('+','')
playstore_df['Size'] = playstore_df['Size'].str.replace(',','')

playstore_df['Size'] = playstore_df['Size'].replace('Varies with device',np.nan)
playstore_df['Size'] = playstore_df['Size'].astype(float)

playstore_df.dropna(inplace=True)
```

```python
sns.scatterplot(x='Size',y='Rating',data=playstore_df)
plt.xlabel('Size(in k)')
plt.ylabel('Rating')
plt.show()
```

**Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.**

```python
playstore_df['Price'] = playstore_df['Price'].str.replace('$','')
playstore_df['Price'] = playstore_df['Price'].astype(float)
```

```python
playstore_df.groupby('Type')['Price'].mean().plot(kind='bar')
plt.xlabel('Type')
plt.ylabel('Average Price')
plt.show()
```
